Plotting User Registration
==========================

```python
from django.db.models import Count
from django.contrib.auth.models import User

# Get the day off of the datetime field as an integer
queryset = User.objects.extra(select={'day': 'date( date_joined )'})

# Values does the equivalent of a GROUP BY
queryset = queryset.values('day')

# Annotate with the count of Users who joined on that day
queryset = queryset.annotate(joined_count=Count('date_joined'))

# Order By the day
queryset = queryset.order_by('day')
```

```python
# Use a list comprehension to grab our annotated value by key
data = [day['joined_count'] for day in queryset]

# Slice to get information about the last month
data = data[:30]

# Plot data in this case a list of single scalar values over time
plot(data)
```

# Matplot lib is capable of some very cool visualizations

```python
%pylab inline

import matplotlib.pyplot as plt
import numpy as np
from numpy import array

xx = np.linspace(-0.75, 1., 100)
n = array([0,1,2,3,4,5])

fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(n, n**2, n**3, color="green", alpha=0.5);
axes[3].set_title("fill_between")
```
